```python
import matplotlib.pyplot as plt
import numpy as np
# Simulated temperature data for Python and Alborz
np.random.seed(0)
num_simulations = 5

Python_temp_data = {
    800:  2534.7,
    1000: 2648.2,
    1200: 2765.1,
    1400: 2887.2,
    1600: 3017.2
}

Alborz_temp_data = {
    800:  2535.23437550659048,
    1000: 2649.100537726586,
    1200: 2766.097681496618,
    1400: 2888.3052071452958,
    1600: 3018.5073680424352
}

# Extract x (temperature) and y (value) data from dictionaries
x_python = list(Python_temp_data.keys())
y_python = list(Python_temp_data.values())

x_alborz = list(Alborz_temp_data.keys())
y_alborz = list(Alborz_temp_data.values())

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_python, y_python, marker='o', label='Python Data')
plt.plot(x_alborz, y_alborz, marker='s', label='Alborz Data')
plt.xlabel('Initial Temperature')
plt.ylabel('Maximum Temperature')
plt.title('Maximum Temperature')
plt.legend()
plt.grid(True)
plt.show()
```
