```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define x-axis
xstart = 0.00201  # Starting at 0 per default
xend = 0.01099  # Ending for example here at 0.1 m = 10 cm
nx = 900  # Number of discretization points - chosen by the user
dx = (xend - xstart) / (nx - 1)  # Grid discretization in the x direction
xvalues = np.linspace(xstart, xend, nx)  # These are the values on the x-axis

# Define target hyperbolic tangent function parameters
ystart = 300.  # Starting value, for example here 300 (K for temperature)
yend = 2100.  # Ending value, for example here 2100 (K for temperature)

# Where to place the flame (as decided by the user)
flamepos = 0.5  # "central flame position" as a percentage of the whole x-axis
xflamepos = xstart + flamepos * (xend - xstart)  # Convert to a real x value
stiff = 3000  # Stiffness parameter for the hyperbolic tangent function - to play with
```

```python
# Compute factor entering the function
fac = stiff * (xvalues - xflamepos)

# Compute hyperbolic tangent and move it between 0 and +1 instead of -1/+1
yth = (1 + np.tanh(fac)) / 2.

# Renormalize with user-defined values
y = ystart + yth * (yend - ystart)

# Plot result
plt.plot(xvalues, y, linewidth=2, color='r')
plt.grid(True)
plt.xlabel('Position along x-axis')
plt.ylabel('Function approximated by hyperbolic tangent')
plt.title('Hyperbolic Tangent Function Approximation')
plt.show()
```
