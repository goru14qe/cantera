```python
import matplotlib.pyplot as plt
import sys
import numpy as np
import cantera as ct
from matplotlib import *
import csv

Temp='1600'
species_names = ['H2', 'H', 'O2', 'OH', 'O', 'H2O', 'HO2', 'H2O2', 'N2']
species_data = np.loadtxt('D:/2_Cantera/Python_files/0D_H2/Alborz_Results/Alborz_Results_H2_'+Temp+'/debug/species_mole_fractions.dat')
```

```python
#   columns         0           1       2          3    4       5   6       7   8
# species_order = [Iteration, Time, Temperature, 'O2', 'H2O','CH4','CO', 'CO2', 'N2']
# Example usage
# Define a function for dynamic plotting

# Define your list of species
species_to_plot = ['Temperature (K)']+species_names

# Create a subplot grid dynamically based on the number of species
num_species = len(species_to_plot)
num_rows = int(np.ceil(num_species / 2))
num_cols = min(2, num_species)
fig, axs = plt.subplots(num_rows, num_cols, figsize=(14, 10))

for i, label in enumerate(species_to_plot):
    row = i // num_cols
    col = i % num_cols
    ax = axs[row, col]  # Get the current subplot

    # Plot your data
    ax.plot(species_data[:, 1], species_data[:, i + 2])

    # Set x-axis ticks format for the current subplot
    ax.ticklabel_format(axis='x', style='sci', scilimits=(0, 0), useMathText=True)

    ax.set_xlabel('Time (s)')
    ax.set_ylabel(label)
    
#plt.tight_layout()
plt.suptitle('Temperature and Mole Fractions for Alborz at '+Temp+'K', fontsize=22)
plt.savefig(f'Results/0D-Alborz_H2_{Temp}.png', dpi=300)
# Show the plot
plt.show()
```
