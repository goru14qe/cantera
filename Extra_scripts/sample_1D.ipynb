```python
import cantera as ct
import numpy as np
import matplotlib.pyplot as plt

gas = ct.Solution('gri30.yaml', 'gri30_mix')
gas.TPX = 300, ct.one_atm, {'CH4':0.9, 'O2':2, 'N2':7.52}

f = ct.FreeFlame(gas,width = 0.03)
f.set_refine_criteria(ratio= 3, slope= 0.07, curve= 0.14)

f.solve(loglevel=1, refine_grid=True, auto=False)

plt.subplots()
plt.suptitle('1D Premixed Laminar Flame of CH4 air mixture (at 300 K and 1 atm initially)')
plt.subplot(2,2,1)
plt.plot(f.grid,f.T)
plt.title('Temperature')
plt.xlabel('Distance from left (m)')
plt.ylabel('Temperature (K)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,2,2)
plt.plot(f.grid,f.velocity)
plt.title('Velocity')
plt.xlabel('Distance from left (m)')
plt.ylabel('Velocity (m/s)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,2,3)
plt.plot(f.grid,f.int_energy)
plt.title('Internal energy')
plt.xlabel('Distance from left (m)')
plt.ylabel('Internal energy (J/kmol)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,2,4)
plt.plot(f.grid,f.heat_release_rate)
plt.title('Heat Release Rate')
plt.xlabel('Distance from left (m)')
plt.ylabel('Heat Release Rate (J/kmol)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplots()
plt.suptitle('Mole Fractions of species in 1D Premixed Laminar Flame of CH4 air mixture')
plt.subplot(2,4,1)
plt.plot(f.grid, f.X[gas.species_index('CH4'),:])
plt.title('CH4')
plt.xlabel('Distance from left (m)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,4,2)
plt.plot(f.grid, f.X[gas.species_index('O2'),:])
plt.title('O2')
plt.xlabel('Distance from left (m)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,4,3)
plt.plot(f.grid, f.X[gas.species_index('N2'),:])
plt.title('N2')
plt.xlabel('Distance from left (m)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,4,4)
plt.plot(f.grid, f.X[gas.species_index('CO2'),:])
plt.title('CO2')
plt.xlabel('Distance from left (m)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,4,5)
plt.plot(f.grid, f.X[gas.species_index('H2O'),:])
plt.title('H2O')
plt.xlabel('Distance from left (m)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,4,6)
plt.plot(f.grid, f.X[gas.species_index('CO'),:])
plt.title('CO')
plt.xlabel('Distance from left (m)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,4,7)
plt.plot(f.grid, f.X[gas.species_index('NO'),:])
plt.title('NO')
plt.xlabel('Distance from left (m)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)

plt.subplot(2,4,8)
plt.plot(f.grid, f.X[gas.species_index('OH'),:])
plt.title('OH')
plt.xlabel('Distance from left (m)')
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', alpha=0.2)


plt.show()
```
